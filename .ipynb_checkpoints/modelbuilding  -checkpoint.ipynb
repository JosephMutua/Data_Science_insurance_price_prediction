```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,7)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Modellingdata.csv')
df.head()
```

```python
len(df['charges'].unique())
```

```python
df1 = df.drop(['Unnamed: 0', 'male', 'children', 'northwest', 'southeast', 'southwest'], axis ='columns')
```

```python
def scatter_plot (x, y):
    plt.scatter (x, y)
    
    
    
    
scatter_plot(df1.age,df1.charges)
```

```python
scatter_plot(df1.bmi,df1.charges)
```

```python
scatter_plot(df1.yes,df1.charges)
```

```python
X = df[['age', 'bmi', 'yes']]
y = df['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2)
```

```python
X_train
```

```python
y_train
```

```python
model = LinearRegression()
model.fit (X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_test, y_test)
```

Please not that the above model is continuos regression model but the model accuracy is low because of the poor dataset provide. This is because other models like random forest and XGboost give better results, but this model is not a classifictaion model

# Modified training data 

```python
df_mod=pd.read_csv("welldist.csv")
df_mod.head()
```

```python
df_mod_smoker_dum= pd.get_dummies(df_mod.smoker)
```

```python
df_joint=pd.concat([df_mod, df_mod_smoker_dum], axis="columns")
```

```python
df_joint.head()
```

```python
df_mod_final=df_joint.drop(["Unnamed: 0","smoker", "no"], axis ="columns")
```

```python
df_mod_final.head()
```

```python
x_mod=df_mod_final[["age","bmi","yes"]]
```

```python
y_mod = df_mod_final.charges
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_mod,y_mod, test_size=0.2)
```

```python
mod_model=LinearRegression()
```

```python
mod_model.fit(X_train,Y_train)
```

```python
mod_model.score(X_test,Y_test)
```
